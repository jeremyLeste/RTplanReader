```python
import dicom
import numpy as np
from stl import mesh
import plotly.offline as py

import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)
```

```python
plan = dicom.read_file("rp.zordan.dcm")
print plan.BeamSequence[0].BlockSequence[0].BlockName
```

```python
print len(plan.BeamSequence[0].BlockSequence[0].BlockData)
print 
bdata= [float(val) for val in plan.BeamSequence[0].BlockSequence[0].BlockData]
x=[]
y=[]

for i in np.arange(0, len(bdata)):
    if i %2  ==0:
        x.append(bdata[i])
    else: 
        y.append(bdata[i])
#print x
print 
#print y
```

```python
trace1 = go.Scatter(
    x = x,
    y = y,
    mode = 'lines', 
    name = 'Block edge', 
    )
data1 = [trace1]
fig1= go.Figure(data=data1)   
py.iplot(fig1) 
```

# STL TOP 

```python
pts_up=[]
for val in x:
    pt=[]
    pt.append(val)
    pt.append(y[x.index(val)])
    pt.append(0)
    pts_up.append(pt)
    
origin1=[0, 0, 0]

stl_top=[]
for i in range(0, len(pts_up)):
    f=[]
    z=i+1
    f.append(pts_up[i])
    if z > len(pts_up)-1: 
        f.append(pts_up[0])
    else:
        f.append(pts_up[z])
    f.append(origin1)
    stl_top.append(f)        
print(stl_top)
```

# STL BOTTOM

```python
pts_down=[]
for val in x:
    pt=[]
    pt.append(val)
    pt.append(y[x.index(val)])
    pt.append(-20)
    pts_down.append(pt)
origin2=[0, 0, -20] 
stl_bot=[]
for i in range(0, len(pts_down)):
    f=[]
    z=i+1
    f.append(pts_down[i])
    if z > len(pts_down)-1: 
        f.append(pts_up[0])
    else:
        f.append(pts_down[z])
    f.append(origin2)
    stl_bot.append(f)
#print stl_bot
```

```python
# Define the 8 vertices of the cube 
vertices = np.array([\
    [-125, -125, -7.5],
    [+125, -125, -7.5],
    [+125, +125, -7.5],
    [-125, +125, -7.5],
    [-125, -125, +7.5],
    [+125, -125, +7.5],
    [+125, +125, +7.5],
    [-125, +125, +7.5]])
# Define the 12 triangles composing the cube
faces = np.array([\
    [0,3,1],
    [1,3,2],
    [0,4,7],
    [0,7,3],
    [4,5,6],
    [4,6,7],
    [5,1,2],
    [5,2,6],
    [2,3,6],
    [3,7,6],
    [0,1,5],
    [0,5,4]])

# Create the mesh
cube = mesh.Mesh(np.zeros(faces.shape[0], dtype=mesh.Mesh.dtype))
for i, f in enumerate(faces):
    for j in range(3):
        cube.vectors[i][j] = vertices[f[j],:]

# Write the mesh to file "cube.stl"
cube.save('appli.stl')
print vertices
print faces
print cube[0]
print cube.vectors[0]
```
